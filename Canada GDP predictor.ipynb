```python
reg = linear_model.LinearRegression()
```

```python
canada_modif_df.columns
```

```python
canada_modif_df.rename(columns={"year ": "year","income":"income"})
```

```python
canada_modif_df.columns
```

```python
canada_modif_df.columns = canada_modif_df.columns.str.replace(" ","")
```

```python
canada_modif_df.columns
```

```python
reg.fit(canada_modif_df[["year"]],canada_modif_df.income)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("GDP income")
plt.title("year vs GDP income")
plt.scatter(canada_modif_df.iloc[:,0],canada_modif_df.iloc[:,1],marker="+",color="blue")
plt.plot(canada_modif_df[["year"]],canada_modif_df.income)
```
